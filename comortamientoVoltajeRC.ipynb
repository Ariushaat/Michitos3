```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores de R, C y V0
R = 1000  # Resistencia en ohmios
C = 0.001  # Capacidad en faradios
V0 = 10  # Diferencia de potencial inicial en voltios

# Definir el rango de tiempo
t = np.linspace(0, 5*R*C, 1000)

# Calcular el voltaje a través del capacitor en función del tiempo
Vc = V0 * (1 - np.exp(-t/(R*C)))

# Graficar
plt.plot(t, Vc, label='Voltaje a través del capacitor')
plt.title('Comportamiento del Voltaje en un Circuito RC')
plt.xlabel('Tiempo (s)')
plt.ylabel('Voltaje (V)')
plt.legend()
plt.grid(True)
plt.show()
```
